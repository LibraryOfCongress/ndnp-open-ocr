```python

import boto3
import os
table_name = 'ndnp-open-ocr-table'
job_id = 'notvalidated__e9f387e0-7214-434a-bc33-412c51531ab7'
os.environ['AWS_PROFILE']='NDNP_OPEN_OCR_DEVELOPER_DEV_profile'
dynamodb = boto3.resource("dynamodb")
table = dynamodb.Table(table_name)
try:
    response = table.query(
        KeyConditionExpression=boto3.dynamodb.conditions.Key('pk').eq('JOB')
    )
    print(response['Items'])
except Exception as e:
    print("Error fetching items from DynamoDB: %s", e)
```
