```python
import boto3
import csv

# Initialize the S3 client
def init_s3_client(aws_profile):
    session = boto3.Session(profile_name="NDNP_OPEN_OCR_DEVELOPER_DEV_profile")
    s3 = session.client('s3')
    return s3

def list_s3_objects(s3, bucket_name, prefix):
    """
    List all objects in an S3 bucket for a given prefix.
    :param s3: The S3 client object
    :param bucket_name: The name of the S3 bucket
    :param prefix: The prefix to list objects for
    :return: A list of keys (object paths)
    """
    keys = []
    paginator = s3.get_paginator('list_objects_v2')
    for page in paginator.paginate(Bucket=bucket_name, Prefix=prefix):
        if 'Contents' in page:
            for obj in page['Contents']:
                keys.append(obj['Key'])
    return keys

def write_to_csv(bucket_name, keys, output_file):
    """
    Write bucket name and keys to a CSV file.
    :param bucket_name: The name of the S3 bucket
    :param keys: A list of keys to write
    :param output_file: The path to the CSV file
    """
    with open(output_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Bucket Name', 'Key'])
        for key in keys:
            writer.writerow([bucket_name, key])

if __name__ == '__main__':
    # Specify the AWS profile name
    aws_profile = "NDNP_OPEN_OCR_DEVELOPER_DEV_profile"

    # Initialize the S3 client
    s3 = init_s3_client(aws_profile)

    # Define the bucket name and prefixes
    bucket_name = 'loc-preservation'
    prefixes = [
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20050906_alpha/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20050927_bravo/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20051020_charlie/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20051102_delta/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20051115_echo/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20051123_foxtrot/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060215_golf/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060215_hotel/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20051209_india/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20051216_kilo/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20051220_lima/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20051229_mike/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20051229_november/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060105_oscar/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060106_papa/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060106_quebec/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060107_romeo/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060116_sierra/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060215_tango/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060215_uniform/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060215_victor/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060215_whiskey/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060215_xray/',
        'loc-preservation/lcbp/ndnp/loc/batch_lc_20060323_yankee/'
    ]

    # Collect all the keys from each prefix
    all_keys = []
    for prefix in prefixes:
        keys = list_s3_objects(s3, bucket_name, prefix)
        all_keys.extend(keys)

    # Output CSV file
    output_file = 'LOC_Preservation_Unfreeze_Keys_091024.csv'

    # Write the collected data to the CSV file
    write_to_csv(bucket_name, all_keys, output_file)
    print(f'CSV file has been generated: {output_file}')
```
