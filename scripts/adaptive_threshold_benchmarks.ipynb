```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import pytesseract
import time

# Function to simulate adaptive preprocessing with OpenCV's adaptiveThreshold
def adaptive_preprocess(image, block_size, C):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, C)
image = cv2.imread("/Volumes/DLP1/batch_dlc_kite_ver18/data/sn83030214/00206531290/1877050301/0021.tif")
# Test cases with different parameter settings for adaptive preprocessing
test_cases = [
    {"block_size": 11, "C": 2},
    {"block_size": 11, "C": 5},
    {"block_size": 15, "C": 2},
    {"block_size": 15, "C": 5},
    {"block_size": 19, "C": 2},
    {"block_size": 19, "C": 5},
    {"block_size": 23, "C": 2},
    {"block_size": 23, "C": 5},
]

results = []

for i, params in enumerate(test_cases):
    start_time = time.time()
    
    preprocessed_image = adaptive_preprocess(image, params["block_size"], params["C"])
    
    ocr_result = pytesseract.image_to_string(preprocessed_image)
    
    word_count = len(ocr_result.split())
    
    time_taken = time.time() - start_time
    
    results.append({
        "Test Case": i + 1,
        "Block Size": params["block_size"],
        "C": params["C"],
        "Word Count": word_count,
        "Time Taken (s)": time_taken
    })
```
